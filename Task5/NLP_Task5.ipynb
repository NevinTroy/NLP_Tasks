```python
import pandas as pd
import os 
df = pd.read_csv(r"D:\Documents\SEM7\NLP\NLP_Tasks\Task5\archive\IMDB Dataset.csv")
```

```python
df.head()
```

```python
#Data preprocessing (remove HTML tags)
df['review'] = df['review'].str.replace(r'<[^<>]*>', '', regex=True)
```

```python
df.head()
```

```python
#Stem the document
from nltk.stem.porter import PorterStemmer
porter = PorterStemmer()

def stemmer(text):
    return [porter.stem(word) for word in text.split()]
```

```python
stemmer("Coders love coding and thus they code")
```

```python
#Vectorization of doc
from sklearn.feature_extraction.text import TfidfVectorizer as TIV
tfidf = TIV(strip_accents = None, lowercase = False, norm = "l2", tokenizer = stemmer,use_idf = True, smooth_idf = True)
```

```python
Y = df.sentiment.values
X = tfidf.fit_transform(df['review'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state = 1, test_size = 0.3, shuffle = False)

import pickle
from sklearn.linear_model import LogisticRegressionCV
logit = LogisticRegressionCV(cv = 5, scoring = 'accuracy', max_iter = 50).fit(X_train, Y_train)
saved_model = open('saved_model.sav', 'wb')
pickle.dump(logit, saved_model)
saved_model.close()
```

```python
filename = 'saved_model.sav'
saved_logit = pickle.load(open(filename, 'rb'))
saved_logit.score(X_test, Y_test)
```
